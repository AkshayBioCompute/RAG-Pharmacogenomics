```python
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_pinecone import PineconeVectorStore
from langchain_cohere import CohereRerank
from langchain.retrievers import BM25Retriever, EnsembleRetriever, ContextualCompressionRetriever
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
pinecone = PineconeVectorStore(embedding=embeddings, index_name="files-for-testing")
vectordb_retriever = pinecone.as_retriever(search_kwargs={"k": 50})
keyword_retriever = BM25Retriever.from_documents(documents)
keyword_retriever.k =  50
ensemble_retriever = EnsembleRetriever(retrievers=[vectordb_retriever,keyword_retriever], weights=[0.5, 0.5])
compressor = CohereRerank()
retriever = ContextualCompressionRetriever(base_compressor=compressor, base_retriever=ensemble_retriever)
```
